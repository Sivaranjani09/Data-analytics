# Model building

```python
x=dfout.drop(['Salary'],axis=1)
x
```

```python
y=dfout['Salary']
y
```

```python
#split x and y into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#data normalization with sklearn
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#fit scaler on training data
x_train=sc.fit_transform(x_train)
#transform testing data
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
###Machine learning techniques
```

```python
#importing decisionTreeRegressor from sklearn tree librarry
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(max_depth=3)
```

```python
dt.fit(x_train,y_train)
```

```python
#score of train model
print(dt.score(x_train,y_train))
#score of test model
print(dt.score(x_test,y_test))
```

```python
#using dt.predict function on x_test and calling it to predict value
y_pred=dt.predict(x_test)
```

```python
#checking predicted value
print(y_pred)
```

```python
from sklearn import metrics
```

```python
#checking r2_square to y_test,y_pred to predict accuracy
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error for decision tree regressor is:',round(100*(r_square)))
```

```python
#using dt.predict function on x_test and calling it to y_predict to predict a value
y_pred=dt.predict(x_test)
#import  metrices from sklearn library
from sklearn import metrics
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error adsociated with decision tree regressor is:',round(100*(r_square),2))
```

# Rework on different models

Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rfr=RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
```

```python
#using dt.predict function on x_test and calling it to y_predict to predict a value
y1_pred=rfr.predict(x_test)
```

```python
#Calculating MAE,MSE,MAPE and accuracy
print('Mean Absolute Error(MAE):',metrics.mean_absolute_error(y_test,y1_pred))
print('Mean Squared Error(MSE):',metrics.mean_squared_error(y_test,y1_pred))
#MAE/MSE are performancce measurers
print('Root Mean Squared Error(RMSE):',np.sqrt(metrics.mean_squared_error(y_test,y1_pred)))
#RMSE is the measure of how weel ML model can perform
mape=np.mean(np.abs((y_test-y1_pred)/np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(1000*(1-mape),2)
print('Mean Absolute Percentage Error(MAPE):',round(mape*100,2))
print('Accuracy',round(100*(1-mape),2))
```

```python
#Cross validation:resampiling method that uses different portions of data to test and train a model on different iterations
from sklearn.model_selection import ShuffleSplit
model=LinearRegression()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

```python
#Cross validation:resampiling method that uses different portions of data to test and train a model on different iterations
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeRegressor(max_depth=3)
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```
